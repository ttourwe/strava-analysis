```python
def activity_to_dict(activity):
# note: we could also retrieve laps
    return {
        'type': activity.type,
        'distance': activity.distance,
        'moving time': activity.moving_time,
        'total elevation gain': activity.total_elevation_gain,
        'start date': activity.start_date,
        'start latitude': activity.start_latitude,
        'start longitude': activity.start_longitude,
        'average speed': activity.average_speed,
        'max speed': activity.max_speed,
        'has heartrate': activity.has_heartrate,
        'average heartrate': activity.average_heartrate,
        'max heartrate': activity.max_heartrate,
        'average temp': activity.average_temp,
        'calories': activity.calories,
        'id': activity.id
    }

overview_df = pd.DataFrame([activity_to_dict(act) for act in activities])
```

```python
preprocess_df = overview_df[['average heartrate','average speed','distance','id','start date', 'type']]
preprocess_df['distance'] = preprocess_df['distance'].astype(float)
preprocess_df = preprocess_df[(preprocess_df['distance'] > 30000) & (preprocess_df['type'] == 'Run')]
```

```python
stream_names = ['latlng','time','altitude','heartrate', 'distance']

df = pd.DataFrame()

for activity_id in preprocess_df['id']:
    stream = client.get_activity_streams(activity_id, types=stream_names)
    series_of_streams = {}
    for s in stream_names:
        if(s == 'latlng'):
            series_of_streams['lat'] = pd.Series([latlng[0] for latlng in stream[s].data])
            series_of_streams['lng'] = pd.Series([latlng[1] for latlng in stream[s].data])
        else: 
            series_of_streams[s] = pd.Series(stream[s].data)
    series_of_streams['id'] = pd.Series([activity_id for _ in range(len(series_of_streams['lat']))])
    df = df.append(pd.DataFrame(series_of_streams))
    
full_df = preprocess_df.merge(df, on='id')
full_df
```

```python
full_df['timestamp'] = full_df['start date'] + pd.to_timedelta(full_df['time'], unit='S')
full_df.sort_values('timestamp')
```

```python
min_distance = 100

def get_std_distances_and_df(training_df):
    std_distances = np.arange(training_df['distance_y'].min(),training_df['distance_y'].max(),min_distance)
    return (std_distances, pd.DataFrame(std_distances, columns=['distance_y']))
    
def prepare_for_interpolation(training_df, std_distances_df):
    training_df['incremental duration'] = (training_df['timestamp'] - training_df['timestamp'].shift(1)).dt.seconds
    training_df['incremental duration'] = training_df['incremental duration'].fillna(0)
    training_df['cumulative duration'] = training_df['incremental duration'].cumsum()
    return training_df.merge(std_distances_df, on='distance_y', how='outer')
    
def interpolate(prepared_df, std_distances):
    interpolated_df = prepared_df.set_index('distance_y').sort_index().interpolate(method='index')
    interpolated_df_with_std_distances = interpolated_df.ix[std_distances]
    first_duration = interpolated_df_with_std_distances.loc[prepared_df['distance_y'].min(),'cumulative duration']
    interpolated_df_with_std_distances['derived duration'] = interpolated_df_with_std_distances['cumulative duration'].diff().fillna(first_duration)
    return interpolated_df_with_std_distances[1:]
    
def standardize(training_df):
    training_df = training_df.copy().sort_index().reset_index()
    std_distances, std_distances_df = get_std_distances_and_df(training_df)
    prepared_df = prepare_for_interpolation(training_df, std_distances_df)
    return interpolate(prepared_df, std_distances)
    
#standardized_data = full_df.set_index('timestamp').sort_index().groupby('start date').apply(standardize).drop(['timestamp'],axis=1).reset_index()
standardized_data = full_df.set_index('timestamp').sort_index().groupby('start date').apply(standardize)[['altitude','heartrate','derived duration','lat','lng']].reset_index()
```
